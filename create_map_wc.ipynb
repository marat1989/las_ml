```python
import pandas as pd
import main
# загружаем wc
water_content_data = pd.read_csv(main.data_dir + "\\water_content.csv", delimiter=';')
water_content_data = water_content_data.rename(columns={'Скважина': main.keys_dict[main.kid_well]})
water_content_data = water_content_data.rename(columns={' Начальная обводненность ': 'wc'})
water_content_data = water_content_data.rename(columns={'Дата запуска': 'date_start'})
del water_content_data['date_start']
water_content_data
```

```python
coord_data = pd.read_csv(main.data_dir + "\coord.csv", delimiter=';')
coord_data = coord_data.rename(columns={'well': 'well_name_uwi'})
coord_data = coord_data.rename(columns={'x': 'X-Coord'})
coord_data = coord_data.rename(columns={'y': 'Y-Coord'})
del coord_data['num']
del coord_data['field']
del coord_data['z']
del coord_data['plastorig']
coord_data.replace(' ', '', regex=True, inplace=True)
#coord_data.replace('314_', '', regex=True, inplace=True)
coord_data['X-Coord'] = coord_data['X-Coord'].astype('float64')
coord_data['Y-Coord'] = coord_data['Y-Coord'].astype('float64')
coord_data
```

```python
well_name_dict = pd.read_csv(main.data_dir + "\well_name_key.csv", delimiter=';')
well_name_dict
```

```python
map_wc = pd.merge(coord_data, well_name_dict, on='well_name_uwi')
map_wc = pd.merge(map_wc, water_content_data, on='well_name')
map_wc.shape
map_wc
del map_wc['well_name']
map_wc.to_csv('map_wc.csv', sep =';', columns = ['well_name_uwi', 'X-Coord', 'Y-Coord', 'wc'], index = False, header =False,
             line_terminator = ';\n')
```
