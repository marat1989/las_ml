```python
import numpy as np
import pandas as pd
import main
las_data = pd.read_csv(main.csv_las_out_full_path, delimiter=';')
```

```python
wells_list = np.unique(las_data['well_name'].values).tolist()
len(wells_list)
```

```python
las_data.head()
```

```python
las_data_cutted = pd.DataFrame()

for well in wells_list:    
    well_las_data = las_data[las_data['well_name'] == well]
    cuted_well_las, not_empty = main.cut_data_frame_by_satur(well_las_data)
    if not_empty:
         las_data_cutted = las_data_cutted.append(cuted_well_las)
print(len(las_data_cutted['well_name'].value_counts().index))
las_data_cutted.describe()
```

```python
# загружаем wc
water_content_data = pd.read_csv(main.data_dir + "\\water_content.csv", delimiter=';')
water_content_data = water_content_data.rename(columns={'Скважина': main.keys_dict[main.kid_well]})
water_content_data = water_content_data.rename(columns={' Начальная обводненность ': 'wc'})
water_content_data = water_content_data.rename(columns={'Дата запуска': 'date_start'})
water_content_data.head()
```

```python
las_data_cutted = pd.merge(las_data_cutted, water_content_data, on=main.keys_dict[main.kid_well])
las_data_cutted.head()
```

```python
coord_data = pd.read_csv(main.data_dir + "\\coord.csv", delimiter=';')
coord_data = coord_data.rename(columns={'num': 'well_name_1'})
coord_data = coord_data.rename(columns={'well': 'UWI'})
del coord_data['field']
del coord_data['plastorig']

las_data_cutted = pd.merge(las_data_cutted, coord_data, on='UWI')
las_data_cutted.head()
```

```python
las_data_cutted['well_name'] = las_data_cutted['well_name_1']

del las_data_cutted['STRT']
del las_data_cutted['STOP']
del las_data_cutted['date_start']
del las_data_cutted['UWI']
del las_data_cutted['well_name_1']
del las_data_cutted['z']

las_data_cutted.info()

```

```python
las_data_cutted.replace(' ', '', regex=True, inplace=True)
las_data_cutted['x'] = las_data_cutted['x'].astype('float64')
las_data_cutted['y'] = las_data_cutted['y'].astype('float64')
las_data_cutted.head()
```

```python
las_data_cutted.to_csv('las_data_cutted.csv', sep=';', index = False)
```
