Загрузка возможных названий кривых

```python
import numpy as np
import pandas as pd
import main

las_header = pd.read_csv('new_las\\las_header.csv', delimiter=';')
del las_header['Не стандартное']
del las_header['Ед. измерения']

las_header = las_header.rename(columns={'Стандартное имя кривой': 'cur_name'})
las_header = las_header.rename(columns={'Описание кривой': 'desc'})

las_header['cur_name'] = las_header['cur_name'].astype(str)
las_header['desc'] = las_header['desc'].astype(str)

print(las_header.count())
las_header.head()
```

```python
las_header = las_header.drop_duplicates(subset = 'cur_name')
print(las_header.count())
las_header.head()
```

```python
las_header['cnt'] = 0
las_header.head()
```

```python
import csv
import lasio

def exist_val_in_list(val, list):
    val_exist = False
    for list_v in list:
        if list_v == val:
            val_exist = True
            break
    return val_exist

def set_cur_name_counts(las_dir, lh):
    las_files = main.work_with_dir.GetFilesInDir(las_dir, main.ext_format)
    print('las_files:', las_files)
    print('las_dir:', las_dir)
    
    i_well = 0
    for las_file_name in las_files:
        dict_list = []
        # инициализируем словарь и заполняем в соответсвии с ласом
        l = lasio.read(las_dir + "\\" + las_file_name)
        well_name = las_file_name[:-4]
        print(well_name)
        # получить список кривых
        for lh_name in lh['cur_name'].values:
            exist_val = exist_val_in_list(lh_name, l.keys())
            if exist_val:
                idx = lh[lh['cur_name']==lh_name].index.values[0]
                lh.ix[idx, 'cnt'] = lh.ix[idx, 'cnt'] + 1
    return
```

```python
las_dir = "new_las\\"

set_cur_name_counts(las_dir, las_header)
print("create_csv_from_las end")
```

```python
def get_list_cur_name_from_las(las_dir, cur_name_list):
    las_files = main.work_with_dir.GetFilesInDir(las_dir, main.ext_format)
    print('las_files:', las_files)
    print('las_dir:', las_dir)
    
    i_well = 0
    for las_file_name in las_files:
        dict_list = []
        # инициализируем словарь и заполняем в соответсвии с ласом
        l = lasio.read(las_dir + "\\" + las_file_name)
        well_name = las_file_name[:-4]
        print(well_name)
        # получить список кривых
        cur_name_list.extend(l.keys())
    return
```

```python
las_dir = "new_las\\"
cur_name_list = []

get_list_cur_name_from_las(las_dir, cur_name_list)
print("create_csv_from_las end")
```
